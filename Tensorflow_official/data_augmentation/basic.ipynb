### Basic augmentation

from: https://tensorflow.google.cn/api_docs/python/tf/image

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

读取文件，并解码为 RGB 图像。

```python
image_raw = tf.io.read_file('image.jpg')
image = tf.image.decode_jpeg(image_raw, channels=3)
```

输出原始图像和增强图像的函数。

```python
def show(img):
    plt.imshow(img)

def show2picture(img1, img2):
    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
```

输出。

```python
show(image)
```

```python
image.get_shape()
```

```python
# 成为 float 才能 resize
float_image = tf.image.convert_image_dtype(image, dtype=tf.float32)
resize = tf.image.resize(float_image, [500, 500])
show2picture(image, resize)
```

```python
resize_with_pad = tf.image.resize_with_pad(float_image, 500, 500)
show2picture(image, resize_with_pad)
```

```python
resize_with_crop_or_pad = tf.image.resize_with_crop_or_pad(float_image, 500, 500)
show2picture(image, resize_with_crop_or_pad)
```

```python
flip_left_right = tf.image.flip_left_right(image)
show2picture(image, flip_left_right)
```

```python
flip_up_down = tf.image.flip_up_down(image)
show2picture(image, flip_up_down)
```

```python
# 3x 饱和度
adjust_saturation = tf.image.adjust_saturation(image, 3)
show2picture(image, adjust_saturation)
```

```python
# 亮度增强 0.5 倍
adjust_brightness = tf.image.adjust_brightness(image, 0.5)
show2picture(image, adjust_brightness)
```

```python
# 转为灰度图，grayscale 的 shape 是 (3840, 5760, 1) 无法直接显示，需 squeeze 转换。
grayscale = tf.image.rgb_to_grayscale(image)
show2picture(image, tf.squeeze(grayscale))
```

```python
rot90 = tf.image.rot90(image)
show2picture(image, rot90)
```

```python
cropped = tf.image.central_crop(image, central_fraction=0.5)
show2picture(image, cropped)
```
